```python
sea_level = pd.read_csv('sea-level_fig-1.csv', header = None)
sea_level_reduced = sea_level.iloc[7:,:4] # Start at important Data, Avoid last column which has many NaN
sea_level_reduced.columns = sea_level.iloc[6,:4] # Name columns

sl = sea_level_reduced.dropna().astype(float) #Remove Null data, cast as float for plot later

ax = sl.plot.scatter('Year','CSIRO - Adjusted sea level (inches)',ylabel = 'Sea Level (inches)', c = 'darkblue')
ax.set_title('Rise in Sealevel')
ax.set_xlim([1860,2020])




```

## Problem 3

Plot the COVID hospitalization number of Arizona since the fist day in the data spreadsheet.
You may use the data from column 'inpatient_beds_used_covid'.

Download the data from [HealthData.gov](https://healthdata.gov/Hospital/COVID-19-Reported-Patient-Impact-and-Hospital-Capa/g62h-syeh)

In the top right of the website, there is a button to export the csv file.

```python
covid_pd = pd.read_csv('Covid-19.csv', parse_dates = ['date']) # Load data from CSV, Turn Date into DateTime Type
```

```python
covid_pd.head() # Visualize Data
```

```python
covid_pd = covid_pd[covid_pd['state'] == 'AZ'] # Extract data for Arizona
covid_pd = covid_pd.sort_values('date')
covid_pd.head()
```

```python
covid_plot = covid_pd.plot.scatter('date', 'inpatient_beds_used_covid')
covid_plot.set_xlabel('Date')
covid_plot.set_ylabel('Number of Cases')
covid_plot.set_title('AZ Covid Hospitalizations (2020 - 2023)')
covid_plot.tick_params(axis='x', labelrotation = 45) # rotate X-lables to avoid overlap
```
